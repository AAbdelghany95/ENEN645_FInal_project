# Unet++

```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms
import segmentation_models_pytorch as smp
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score

# ------------------------
# Configuration
# ------------------------
DATA_PATH = "deepglope/archive/train"
IMG_SIZE = (512, 512)  # Using 512x512 resolution
BATCH_SIZE = 10
NUM_EPOCHS = 30
NUM_CLASSES = 7
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
GRAD_ACCUM_STEPS = 2
PATIENCE = 3

# Class information
CLASS_NAMES = ['urban_land', 'agriculture_land', 'rangeland', 
               'forest_land', 'water', 'barren_land', 'unknown']
IGNORE_INDEX = 6  # Index of 'unknown' class to ignore in metrics

# ------------------------
# Dataset Class
# ------------------------
class DeepGlobeDataset(Dataset):
    def __init__(self, root_dir):
        if not os.path.exists(root_dir):
            raise ValueError(f"Directory not found: {root_dir}")
        
        self.image_paths = sorted(glob.glob(os.path.join(root_dir, "*_sat.jpg")))
        if not self.image_paths:
            raise ValueError(f"No images found in {root_dir}")
        
        self.mask_paths = [p.replace("_sat.jpg", "_mask.png") for p in self.image_paths]
        self.valid_pairs = [(img, mask) for img, mask in zip(self.image_paths, self.mask_paths) 
                          if os.path.exists(mask)]
        
        if not self.valid_pairs:
            raise ValueError("No valid image-mask pairs found!")
        
        print(f"Found {len(self.valid_pairs)} valid image-mask pairs")
        
        self.img_transform = transforms.Compose([
            transforms.Resize(IMG_SIZE),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                                 std=[0.229, 0.224, 0.225])
        ])
        
        self.mask_transform = transforms.Compose([
            transforms.Resize(IMG_SIZE, interpolation=transforms.InterpolationMode.NEAREST),
        ])

    def __len__(self):
        return len(self.valid_pairs)

    def __getitem__(self, idx):
        img_path, mask_path = self.valid_pairs[idx]
        
        img = Image.open(img_path).convert("RGB")
        img = self.img_transform(img)
        
        mask = np.array(Image.open(mask_path))
        mask = self._rgb_to_class(mask)
        mask = Image.fromarray(mask)
        mask = self.mask_transform(mask)
        mask = torch.from_numpy(np.array(mask)).long()
        
        return img, mask

    def _rgb_to_class(self, mask_rgb):
        class_mask = np.zeros(mask_rgb.shape[:2], dtype=np.uint8)
        color_mappings = [
            ([0, 255, 255], 0),
            ([255, 255, 0], 1),
            ([255, 0, 255], 2),
            ([0, 255, 0], 3),
            ([0, 0, 255], 4),
            ([255, 255, 255], 5),
            ([0, 0, 0], 6)
        ]
        
        for color, class_idx in color_mappings:
            mask = (mask_rgb[:,:,0] == color[0]) & \
                   (mask_rgb[:,:,1] == color[1]) & \
                   (mask_rgb[:,:,2] == color[2])
            class_mask[mask] = class_idx
            
        return class_mask

# ------------------------
# Model Definition (Using Unet++)
# ------------------------
class DeepGlobeModel(nn.Module):
    def __init__(self):
        super().__init__()
        # Create a Unet++ model with a ResNet50 encoder pre-trained on ImageNet.
        self.model = smp.UnetPlusPlus(
            encoder_name="resnet50",        # encoder: resnet50
            encoder_weights="imagenet",       # use pretrained weights for encoder
            in_channels=3,                    # input channels (RGB)
            classes=NUM_CLASSES,              # number of segmentation classes
        )
        
    def forward(self, x):
        # The model automatically upsamples to the input resolution.
        return self.model(x)

# ------------------------
# Training Utilities
# ------------------------
class Trainer:
    def __init__(self, model, train_loader, val_loader):
        self.model = model.to(DEVICE)
        self.train_loader = train_loader
        self.val_loader = val_loader
        
        # AMP initialization (if using CUDA)
        self.scaler = torch.amp.GradScaler()
        
        self.optimizer = optim.Adam(model.parameters(), lr=1e-4)
        self.scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            self.optimizer, 'min', patience=3
        )
        self.criterion = nn.CrossEntropyLoss(ignore_index=IGNORE_INDEX)
        
        # Tracking
        self.best_val_loss = float('inf')
        self.best_iou = 0.0
        self.early_stop_counter = 0
        
        # Metrics storage
        self.train_losses = []
        self.val_losses = []
        self.val_ious = []
        self.class_ious = {name: [] for name in CLASS_NAMES}

    def train_epoch(self, epoch):
        self.model.train()
        total_loss = 0.0
        self.optimizer.zero_grad()
        
        with tqdm(self.train_loader, unit="batch") as tepoch:
            for batch_idx, (images, masks) in enumerate(tepoch):
                if batch_idx % 10 == 0:
                    torch.cuda.empty_cache()
                
                images = images.to(DEVICE)
                masks = masks.to(DEVICE)
                
                if DEVICE.type == 'cuda':
                    with torch.amp.autocast(device_type='cuda'):
                        outputs = self.model(images)
                        loss = self.criterion(outputs, masks) / GRAD_ACCUM_STEPS
                    
                    self.scaler.scale(loss).backward()
                    
                    if (batch_idx + 1) % GRAD_ACCUM_STEPS == 0 or (batch_idx + 1) == len(self.train_loader):
                        self.scaler.step(self.optimizer)
                        self.scaler.update()
                        self.optimizer.zero_grad()
                else:
                    outputs = self.model(images)
                    loss = self.criterion(outputs, masks) / GRAD_ACCUM_STEPS
                    loss.backward()
                    
                    if (batch_idx + 1) % GRAD_ACCUM_STEPS == 0 or (batch_idx + 1) == len(self.train_loader):
                        self.optimizer.step()
                        self.optimizer.zero_grad()
                
                total_loss += loss.item() * GRAD_ACCUM_STEPS
                tepoch.set_postfix(loss=loss.item() * GRAD_ACCUM_STEPS)
        
        epoch_loss = total_loss / len(self.train_loader)
        self.train_losses.append(epoch_loss)
        return epoch_loss

    def validate(self):
        self.model.eval()
        val_loss = 0.0
        all_preds = []
        all_masks = []
        
        with torch.no_grad():
            for images, masks in self.val_loader:
                images = images.to(DEVICE)
                masks = masks.to(DEVICE)
                
                if DEVICE.type == 'cuda':
                    with torch.amp.autocast(device_type='cuda'):
                        outputs = self.model(images)
                        loss = self.criterion(outputs, masks)
                else:
                    outputs = self.model(images)
                    loss = self.criterion(outputs, masks)
                
                val_loss += loss.item() * images.size(0)
                
                preds = torch.argmax(outputs, dim=1).cpu().numpy()
                masks_np = masks.cpu().numpy()
                
                valid_mask = masks_np != IGNORE_INDEX
                valid_preds = preds[valid_mask]
                valid_masks = masks_np[valid_mask]
                
                all_preds.extend(valid_preds.flatten())
                all_masks.extend(valid_masks.flatten())
                
                del images, masks, outputs
                torch.cuda.empty_cache()
        
        avg_val_loss = val_loss / len(self.val_loader.dataset)
        self.val_losses.append(avg_val_loss)
        
        if len(all_preds) > 0:
            precision = precision_score(all_masks, all_preds, average='macro', zero_division=0)
            recall = recall_score(all_masks, all_preds, average='macro', zero_division=0)
            iou_per_class = self._calculate_iou(np.array(all_preds), np.array(all_masks))
            mean_iou = np.nanmean([iou for iou in iou_per_class.values() if not np.isnan(iou)])
            
            self.val_ious.append(mean_iou)
            for i, class_name in enumerate(CLASS_NAMES):
                if i == IGNORE_INDEX:
                    continue
                self.class_ious[class_name].append(iou_per_class[i])
            
            print(f"\nValidation Metrics:")
            print(f"Loss: {avg_val_loss:.4f} | Precision: {precision:.4f} | Recall: {recall:.4f}")
            print("Class-wise IoU:")
            for i, class_name in enumerate(CLASS_NAMES):
                if i == IGNORE_INDEX:
                    continue
                print(f"  {class_name}: {iou_per_class[i]:.4f}")
            print(f"Mean IoU: {mean_iou:.4f}")
            
            if mean_iou > self.best_iou:
                self.best_iou = mean_iou
                self.best_val_loss = avg_val_loss
                torch.save(self.model.state_dict(), "best_model.pth")
                print("Saved new best model!")
                self.early_stop_counter = 0
            else:
                self.early_stop_counter += 1
        else:
            print("Warning: No valid predictions for metric calculation")
            mean_iou = 0.0
            self.early_stop_counter += 1
        
        self.scheduler.step(avg_val_loss)
        return avg_val_loss

    def _calculate_iou(self, preds, labels):
        ious = {}
        for cls in range(NUM_CLASSES):
            if cls == IGNORE_INDEX:
                continue
            pred_inds = (preds == cls)
            target_inds = (labels == cls)
            intersection = (pred_inds & target_inds).sum()
            union = (pred_inds | target_inds).sum()
            ious[cls] = float(intersection) / float(union) if union > 0 else float('nan')
        return ious

    def plot_sample_predictions(self, dataset, num_samples=3):
        self.model.eval()
        indices = np.random.choice(len(dataset), num_samples)
        
        plt.figure(figsize=(15, 5*num_samples))
        for i, idx in enumerate(indices):
            image, mask = dataset[idx]
            with torch.no_grad():
                if DEVICE.type == 'cuda':
                   with torch.amp.autocast(device_type='cuda'):
                        output = self.model(image.unsqueeze(0).to(DEVICE))
                else:
                    output = self.model(image.unsqueeze(0).to(DEVICE))
                pred = torch.argmax(output, dim=1).squeeze().cpu().numpy()
            
            img_np = image.numpy().transpose(1, 2, 0)
            img_np = img_np * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
            img_np = np.clip(img_np, 0, 1)
            
            plt.subplot(num_samples, 3, i*3 + 1)
            plt.imshow(img_np)
            plt.title(f"Sample {i+1} - Input")
            plt.axis('off')
            
            plt.subplot(num_samples, 3, i*3 + 2)
            plt.imshow(mask.numpy(), vmin=0, vmax=NUM_CLASSES-1)
            plt.title("Ground Truth")
            plt.axis('off')
            
            plt.subplot(num_samples, 3, i*3 + 3)
            plt.imshow(pred, vmin=0, vmax=NUM_CLASSES-1)
            plt.title("Prediction")
            plt.axis('off')
        
        plt.tight_layout()
        plt.show()

    def plot_training_metrics(self):
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.plot(self.train_losses, label='Train')
        plt.plot(self.val_losses, label='Validation')
        plt.title("Loss")
        plt.xlabel("Epoch")
        plt.legend()
        
        plt.subplot(1, 3, 2)
        plt.plot(self.val_ious)
        plt.title("Mean IoU")
        plt.xlabel("Epoch")
        
        plt.subplot(1, 3, 3)
        for cls in CLASS_NAMES:
            if cls != 'unknown':
                plt.plot(self.class_ious[cls], label=cls)
        plt.title("Class-wise IoU")
        plt.xlabel("Epoch")
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        
        plt.tight_layout()
        plt.show()

    def evaluate_test_set(self, test_loader):
        self.model.load_state_dict(torch.load("best_model.pth"))
        self.model.eval()
        
        all_preds = []
        all_masks = []
        
        with torch.no_grad():
            for images, masks in test_loader:
                images = images.to(DEVICE)
                masks = masks.to(DEVICE)
                
                if DEVICE.type == 'cuda':
                    with torch.amp.autocast(device_type='cuda'):
                        outputs = self.model(images)
                else:
                    outputs = self.model(images)
                
                preds = torch.argmax(outputs, dim=1).cpu().numpy()
                masks_np = masks.cpu().numpy()
                
                valid_mask = masks_np != IGNORE_INDEX
                valid_preds = preds[valid_mask]
                valid_masks = masks_np[valid_mask]
                
                all_preds.extend(valid_preds.flatten())
                all_masks.extend(valid_masks.flatten())
        
        if len(all_preds) == 0:
            print("Warning: No valid predictions in test set")
            return
        
        precision = precision_score(all_masks, all_preds, average='macro', zero_division=0)
        recall = recall_score(all_masks, all_preds, average='macro', zero_division=0)
        iou_per_class = self._calculate_iou(np.array(all_preds), np.array(all_masks))
        
        print("\nTest Set Evaluation:")
        print(f"Precision: {precision:.4f} | Recall: {recall:.4f}")
        print("Class-wise IoU:")
        for i, class_name in enumerate(CLASS_NAMES):
            if class_name != 'unknown':
                print(f"  {class_name}: {iou_per_class[i]:.4f}")
        
        plt.figure(figsize=(10, 8))
        cm = confusion_matrix(all_masks, all_preds, labels=range(NUM_CLASSES), normalize='true')
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CLASS_NAMES)
        disp.plot(cmap='Blues', values_format='.2f')
        plt.title("Normalized Confusion Matrix")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# ------------------------
# Main Execution
# ------------------------
def main():
    try:
        print("Loading dataset...")
        dataset = DeepGlobeDataset(DATA_PATH)
        
        sample_img, sample_mask = dataset[0]
        print(f"\nSample image size: {sample_img.shape}")
        print(f"Sample mask size: {sample_mask.shape}")
        print(f"Unique mask values: {torch.unique(sample_mask)}")
        
        train_size = int(0.7 * len(dataset))
        val_size = int(0.15 * len(dataset))
        test_size = len(dataset) - train_size - val_size
        
        train_dataset, val_dataset, test_dataset = random_split(
            dataset, [train_size, val_size, test_size],
            generator=torch.Generator().manual_seed(42)
        )
        
        print(f"\nDataset split:")
        print(f"Train: {len(train_dataset)} samples")
        print(f"Val: {len(val_dataset)} samples")
        print(f"Test: {len(test_dataset)} samples")
        
        train_loader = DataLoader(
            train_dataset,
            batch_size=BATCH_SIZE,
            shuffle=True,
            num_workers=2,
            pin_memory=False
        )
        
        val_loader = DataLoader(
            val_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=2,
            pin_memory=False
        )
        
        test_loader = DataLoader(
            test_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=2,
            pin_memory=False
        )
        
        print("\nInitializing model...")
        model = DeepGlobeModel()
        print(f"Using device: {DEVICE}")
        print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
        
        trainer = Trainer(model, train_loader, val_loader)
        print("\nStarting training...")
        
        for epoch in range(NUM_EPOCHS):
            train_loss = trainer.train_epoch(epoch)
            val_loss = trainer.validate()
            
            print(f"\nEpoch {epoch+1}/{NUM_EPOCHS}")
            print(f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
            
            if (epoch+1) % 5 == 0:
                print("\nSample predictions:")
                trainer.plot_sample_predictions(val_dataset)
            
            if trainer.early_stop_counter >= PATIENCE:
                print(f"\nEarly stopping triggered at epoch {epoch+1}")
                print(f"Best validation IoU: {trainer.best_iou:.4f}")
                break
            
            torch.cuda.empty_cache()
        
        print("\nTraining metrics:")
        trainer.plot_training_metrics()
        
        print("\nEvaluating on test set...")
        trainer.evaluate_test_set(test_loader)
        
        print("\nTraining and evaluation completed!")

    except Exception as e:
        print(f"\nError: {str(e)}")
        torch.cuda.empty_cache()

if __name__ == "__main__":
    main()
    torch.cuda.empty_cache()
```
